```python
from CNNEvaluation import CatImageClassifier

import numpy as np

classifier = CatImageClassifier('File_cats.xlsx', './images/')
```

```python
classifier.build_model()
```

```python
classifier.train_model(epochs=100, batch_size=10)
```

```python
classifier.evaluate_model()
classifier.model.save('cat_classifier_model.h5')
np.save('label_encoder_classes.npy', classifier.label_encoder.classes_)
```

```python
classifier.plot_class_distribution()
```

```python
new_image_path = './images/Prova1.png'
predicted_cat_name = classifier.predict_cat(new_image_path)

print(f'The predicted cat name is: {predicted_cat_name}')
```

```python
classifier.plot_predictions(num_samples=5)
```
